```python
## try implement plotting precision recall curve
import matplotlib.pyplot as plt
import numpy as np

preds = np.random.rand(100)
gts = np.random.randint(0, 2, 100)

def plot_pr_curve(preds, gts):
    precisions = []
    recalls = []
    thresholds = np.linspace(0, 1, 100)

    for threshold in thresholds:
        tp = np.sum((preds >= threshold) & (gts == 1))
        fp = np.sum((preds >= threshold) & (gts == 0))
        fn = np.sum((preds < threshold) & (gts == 1))

        precision = tp / (tp + fp) if tp + fp > 0 else 1
        recall = tp / (tp + fn) if tp + fn > 0 else 0
        precisions.append(precision)
        recalls.append(recall)

    plt.figure(figsize=(10, 5))
    plt.plot(recalls, precisions, label='PR Curve')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title('Precision-Recall Curve')
    plt.legend()
    plt.show()

plot_pr_curve(preds, gts)




```
